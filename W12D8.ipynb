```python
import os

# Percorso principale dove si trovano le cartelle
percorso = r"C:\Users\Matteo Perrotta\Downloads\covid-19-data-master\covid-19-data-master\public\data"

# Elenco delle cartelle
nome_cartella = [
    "jhu", "latest", "testing", "vaccinations", "internal_new",
    "internal", "cases_deaths", "excess_mortality", "hospitalizations", "archived"
]

def get_folder_size(path):
    total_size = 0
    for dirpath, _, filenames in os.walk(path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            if os.path.isfile(fp):
                total_size += os.path.getsize(fp)
    return total_size / (1024 * 1024)  # dimensione in MB

# Stampa la dimensione per ogni cartella
for cartella in nomi_cartelle:
    full_path = os.path.join(percorso, cartella)
    size_mb = get_folder_size(full_path)
    print(f"{cartella}: {size_mb:.2f} MB")
```

```python
import os

# Percorso principale
base_path = r"C:\Users\Matteo Perrotta\Downloads\covid-19-data-master\covid-19-data-master\public\data"

# Nomi delle 10 cartelle da analizzare
nomi_cartelle = [
    "jhu", "latest", "testing", "vaccinations", "internal_new",
    "internal", "cases_deaths", "excess_mortality", "hospitalizations", "archived"
]

# Estensioni considerate come metadati
ext_metadati = [".csv", ".json", ".xml"]

# Conta metadati per ciascuna cartella
for cartella in nomi_cartelle:
    full_path = os.path.join(base_path, cartella)
    count = 0
    for root, _, files in os.walk(full_path):
        for file in files:
            if any(file.lower().endswith(ext) for ext in ext_metadati):
                count += 1
    print(f"{cartella}: {count} file metadati")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carica il CSV (se non già caricato)
db = pd.read_csv(r"C:\Users\Matteo Perrotta\Downloads\covid-19-data-master\covid-19-data-master\public\data\hospitalizations\covid-hospitalizations.csv")

# Filtro per le nazioni 
nazioni = ['Italy', 'Germany', 'France']
db_nazioni = db[db['entity'].isin(nazioni)]

# Filtro per le date 
db_periodo = db_nazioni[
    (db_nazioni['date'] >= '2022-05-01') &
    (db_nazioni['date'] <= '2023-04-30')
]

# Filtra per i dati ICU
db_icu = db_periodo[db_periodo['indicator'] == 'Daily ICU occupancy']

# Rimuovi valori NaN
db_icu_clean = db_icu.dropna(subset=['value'])

# Crea il grafico
plt.figure(figsize=(10, 6))
sns.boxplot(data=db_icu_clean, x='entity', y='value')
plt.title("Terapia Intensiva (ICU) - Italia, Germania, Francia\nMaggio 2022 ad Aprile 2023")
plt.xlabel("Nazione")
plt.ylabel("Pazienti in terapia intensiva (ICU)")
plt.grid()
```

Abbiamo confrontanto i dati e realizzato un boxplot con i numeri giornalieri di persone in terapia intensiva nelle 3 seguenti nazioni: France, Germany e Italy.

Possiamo notare che gli ospedali della Germania, in questo frangente di tempo ha avuto un numero maggiore di positivi in terapia intensiva. L'Italia invece in quel periodo ha numeri inferiori rispetto alle altre nazioni.
