```python
# VISUALIZZIAMO L'ANDAMENTO DI MICROSOFT E IBM NEGLI ANNI 2012 E 2013

import pandas as pd
import matplotlib.pyplot as plt #Modulo per grafici e stilizzare grafici 


#Abbiamo dichiarato una variabile db['Date'] e abbiamo reso i dati str nella colonna Date in formato datetime
db['Date'] = pd.to_datetime(db['Date']) 

#Abbiamo dichiarato una variabile Filtro per usare i dati in datetime per prenderci solo l'anno.
#Mi prendera' tutti i db['Date'] >= 2012 e <= 2013
filtro = (db['Date'].dt.year >= 2012) & (db['Date'].dt.year <= 2013)

#Creo una variabile dati_filtrati che contenga db['Filtro'] cosÃ¬ da darci solo i dati che ci servono
dati_filtrati = db[filtro]

#Crea una figura su cui disegnare il grafico, 12 pollici in larghezza e 6 in altezza
plt.figure(figsize=(12, 6)) 

#Traccia una linea per i Dati Filtrati della colonna Date, prendendo escluviamente la colonna MSFT;
#il nome della linea sarÃ  Microsoft (MSFT) e sara' di colore blu
plt.plot(dati_filtrati['Date'], dati_filtrati['MSFT'], label='Microsoft (MSFT)', color='blue')

#Traccia una linea per i Dati Filtrati della colonna Date, prendendo escluviamente la colonna IBM;
#il nome della linea sarÃ  IBM e sara' di colore rosso
plt.plot(dati_filtrati['Date'], dati_filtrati['IBM'], label='IBM', color='red')

#Abbiamo creato una leggenda, i cui nomi sono indicati dal parametro label = ...
plt.legend()

#Abbiamo creato una griglia con linee verticali e orizzontali per facilitarne la comprensione
plt.grid()
```

```python
#Delle band che suonano sia Progressive che Thrash, realizziamo un barplot che mostri quanti fan ci sono in totale rispetto al paese di origine 
#(nota: in questo caso specifico, per accedere alla colonna "style" dovremo usare .loc[:, "style"] e non .style perchÃ© quest'ultimo Ã¨ un metodo)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

condizione = (
    db.loc[:,'style'].str.contains('Progressive', case=False, na=False) &
     db.loc[:,'style'].str.contains('Thrash', case=False, na=False)
)

band_filtrate = db[condizione]

fan_paese = band_filtrate.groupby('origin')['fans'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(data=fan_paese, x='origin', y='fans')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Visualizziamo un istogramma delle date di formazione delle band 
# (potremo dover convertire in qualche modo questo dato); si nota qualche trend?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Rimuovi le righe con 'formed' mancante
db = db.dropna(subset=['formed'])

# Estrai lâ€™anno (come intero)
db['year_formed'] = db['formed'].astype(int)

# Visualizzazione con istogramma
plt.figure(figsize=(8, 5))
sns.histplot(data=db, x='year_formed', bins=10, color='darkred')
plt.title('Distribuzione delle date di formazione delle band')
plt.xlabel('Anno di formazione')
plt.ylabel('Numero di band')
plt.tight_layout()
plt.show()
```

```python
# CON UN GRAFICO A BARRE CONFRONTIAMO I VOTI TOTALI PRESI DAI TRE CANDIDATI (COME SOMMA DI TUTTI I DISTRETTI)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Somma dei voti per ciascun candidato
c1 = db['Coderre'].sum()
c2 = db['Bergeron'].sum()
c3 = db['Joly'].sum()

# Creo un DataFrame con i totali
totali = pd.DataFrame({
    'Candidato': ['Coderre', 'Bergeron', 'Joly'],
    'Voti': [c1, c2, c3]
})

# Grafico a barre
plt.figure(figsize=(8, 5))
sns.barplot(data=totali, x='Candidato', y='Voti')
plt.title('Voti totali per ciascun candidato')
plt.ylabel('Numero di voti')
plt.grid()
```

```python
#Con un grafico a barre confrontiamo il numero di votanti per i primi dieci distretti

dati_necessari = db.head(10)

grafico = pd.DataFrame({
    'Distretto': dati_necessari['district'],
    'Votanti': dati_necessari['total'],
})
    
plt.figure(figsize=(8, 5))
sns.barplot(data=grafico, x='Distretto', y='Votanti')
plt.title('Numero di votanti nei primi 10 distretti')
plt.xlabel('Distretto')
plt.ylabel('Numero di votanti')
plt.xticks(rotation=45)
```
